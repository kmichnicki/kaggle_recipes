# Classification

What are the different strategies we could use for classification? There are a lot. <a href='https://en.wikipedia.org/wiki/Category:Classification_algorithms'>Wikipedia</a> has a list of over 20. 

1. Vanilla neural networks. 
2. Bayesian classifiers.
3. Random forest. 

Let's state the problem. Given a list of ingredients, we need to pick one cuisine out of 20. So the algorithm gets ingredients and spits out a prediction. 

One simple idea is assign a conditional probability for a particular ingredient showing up given the cuisine. We then use bayes rule to update our belief about the cuisine given a prior distribution that matches the frequency of each cuisine in the training data. Then the product that is the highest wins. One problem with this approach is that there may be a new ingredient that we've never seen and we don't want to say that the probability is zero. We would just omit it. What if an ingredient shows up for a cuisine and all of the ingredients look brazilian except there is one ingredient that has never shown up in brazilian cuisine, since our brazilian sample is small this is likely, and yet it shows up in a different cuisine which otherwise is a poor choice. We wouldn't want to throw brazilian out as a candidate just because of one ingredient. Or would we? At the very least, we would want some way of not throwing away possibilities because of a single discrepency, especially if the competing candidates are bad. 

A problem with this simple bayesian update is that it ignores pairs of data. Maybe if flour is in the list of ingredients, sugar will be there too and if sugar is not there, it is much more information than if it were not there and we didn't know if flour was there. However considering 3, 4 and higher body effects would be computationally prohibative. 

Ok, back to earth. What are some other pieces of information that are relavent? 

1. Each recipe has a different number of ingredients.
2. The order of the ingredients may matter. 
3. Pairs, triplets, etc... may matter.
4. Cuisines happen with different frequencies. 
5. How about length of the ingredients for a particular cuisine? 
6. Frequency of letters?

Candidates:

1. Neural networks, vanilla.

I want the output to be 20 real numbers representing the confidence that the ith cuisine is the correct one. With each additional ingredient I want the confidence to change. I want the thing to have a rough memory of previous ingredients in case many-body data matters and in case order matters. Bayesian update satisfies two of these, it doesn't care about order, but recursive neural networks care about all of these. However, number is not something that it cares about. ## Does order matter?

I have a belief that word order matters. How do I test this? Well at a first pass, I can just list the order of ingredients in different recipes to see if there is a difference. 

```python
def construct_pairs(ingredients):
    ingredients2=[i for i in ingredients]
    output=[]
    for i,v1 in enumerate(ingredients):
        for j,v2 in enumerate(ingredients2):
            if i<j:
                output.append(v1+str(",")+v2)
    return output

paired_data=[]
for recipe in data:
    paired_data.append({'cuisine':recipe['cuisine'],'id':recipe['id'],'ingredients':construct_pairs(recipe['ingredients'])})
    
paired_ingredients=Counter()
for recipe in paired_data:
    for ingredient in recipe['ingredients']:
        paired_ingredients[ingredient]+=1
        
print "There are "+str(len(paired_ingredients))+" paired ingredients total.\n\n"

counts=Counter(paired_ingredients)
n=10
print "The following is a list of the "+str(n)+" most common pairs of ingredients.\n"
for key,val in counts.most_common(n):
    print str(key) + " "*(30-len(key)) + " : " + str(val)
```

Indeed, my hypothesis is correct. For instance, "salt, onions" happens about 50% more than "onions,salt". The real question though is whether or not we can use this to make any predictions about specific cuisine? To get a handle on this, I'll just print out the most popular pairs per cuisine. 



```python
def get_paired_ingredients_of_cuisine(cuisine):
    ingredients=Counter()
    
    for recipe in paired_data:
        if recipe['cuisine']==cuisine:
            for ingredient in recipe['ingredients']:
                ingredients[ingredient]+=1
    
    return ingredients
```

```python
cuisine_ingredients_paired={}
for cuisine in cuisines:
    cuisine_ingredients_paired[cuisine]=get_paired_ingredients_of_cuisine(cuisine)
    
```

```python
n=10
for cuisine,_ in cuisines.most_common():
    print cuisine
    for key,val in cuisine_ingredients_paired[cuisine].most_common(n):
        print "\t"+str(key)+" : "+str(val)
    print "\n"
```

So it looks like a lot of recipes have similar preferences for word order. For instance, salt seems to come before onions pretty consistently. However, if we were to see onions come before salt, then what? Which cuisine is most likely to do that? To answer this we need to know something about the relative frequency of the different orderings. I'm going to test this out on number(salt,onions)/number(onions,salt) since that is the most popular pair of ingredients.

In what follows, I calculate a function N/M where N and M are integers. N is the number of times that 'salt,onions' happens for that cuisine and M is the number of times that 'onions,salt' happens. I calculate the ration N/M but I also calculate the error by propagating errors. In general I know that the percent error for a function $f(N,M)$ is just $\frac{\sqrt{\sigma_N^2 \frac{\partial f}{\partial N}^2 + \sigma_M^2 \frac{\partial f}{\partial M}^2}}{f(N,M)}$. For the function f=N/M the percent error is $\sqrt{\frac{\sigma_N^2}{N^2}+\frac{\sigma_M^2}{M^2}}$ so the total error is just $\frac{N}{M}\sqrt{\frac{\sigma_N^2}{N^2}+\frac{\sigma_M^2}{M^2}}$.

```python
import math

for cuisine,_ in cuisines.most_common():
    print cuisine
    x=cuisine_ingredients_paired[cuisine]
    if x['onions,salt']!=0 and x['salt,onions']!=0:
        print "num('salt,onions')="+str(x['salt,onions'])
        print "num('onions,salt')="+str(x['onions,salt'])
        value=1.0*x['salt,onions']/x['onions,salt']
        percent_error = math.sqrt((1/x['onions,salt']+1.0/x['salt,onions']))
        error = value*percent_error
        print "num('salt,onions')/num('onions,salt')="+ str(value)+ " +/- "+ str(error)
    else:
        print "One of the combinations does not happen."
    print "\n"
```

On the low end, russian cuisine lists salt,onions 1.19 times more than onions,salt. Thai on the other hand does it 2.63 times more. Is this just a statistical fluke? Russian cuisines has an error of 0.16 while thai cuisine has an error of about 0.49. So even when we take into account the error, we find a range of .79 between them, which is almost two thai standard deviations apart. Ok, but we have 20 examples. With 20 examples and random sampling, we may be very likely to find a pair that is so far apart. Could this just be something like p-hacking? Looking at consecutive pairs of cuisines, the ratios are out a standard deviation of each other 13 out of 19 times. 

This needs further investigation however, the evidence so far indicates that the hypothesis "order matters" is more likely than "order does not matter.

## Do pairs, triplets, etc... matter?

How do I formalize this? Well, I just look at the frequency of pairs of ingredients. I did this in the ordered case above. I found that certain pairs were more likely found in some cuisines. Like "sugar, fish sauce" indicates thai or vietnamese. However, fish sauce already did this. The question is whether the combination does a better job? How do I test this? Well for each cuisine I can calculate the proobability of seeing that pair, call that $p_c(x,y)$ where $x$ and $y$ are the two ingredients and $c$ denotes the cuisine. I can measure how often these differ between cuisines. So average over all pairs of cuisines and all pairs of ingredients? 

$measure=\frac{1}{\#pairs}\frac{1}{N^2-N} \sum_{(x,y),c,c'} |p_c(x,y)-p_{c'}(x,y)|$

First I need a function that gives me $p_c(x,y)$.

```python
def construct_pairs_unordered(ingredients):
    ingredients2=[i for i in ingredients]
    output=[]
    for i,v1 in enumerate(ingredients):
        for j,v2 in enumerate(ingredients2):
            if i!=j:
                output.append(v1+str("----")+v2)
    return output

paired_data_unordered=[]
for recipe in data:
    paired_data_unordered.append({'cuisine':recipe['cuisine'],'id':recipe['id'],'ingredients':construct_pairs_unordered(recipe['ingredients'])})
    

        
```

```python



def get_paired_ingredients_of_cuisine_unordered(cuisine):
    ingredients=Counter()
    
    for recipe in paired_data_unordered:
        if recipe['cuisine']==cuisine:
            for ingredient in recipe['ingredients']:
                ingredients[ingredient]+=1
    
    return ingredients

cuisine_ingredients_paired_unordered={}
for cuisine in cuisines:
    cuisine_ingredients_paired_unordered[cuisine]=get_paired_ingredients_of_cuisine_unordered(cuisine)

num_paired_ingredients_by_cuisine_unordered=Counter()
for cuisine in cuisines:
    length=0
    for a_pair in cuisine_ingredients_paired_unordered[cuisine]:
        length+=cuisine_ingredients_paired_unordered[cuisine][a_pair]
    
num_paired_ingredients_by_cuisine_unordered[cuisine]=length

def p_pair(cuisine,pair):
    value=0
    length=0
    length=num_paired_ingredients_by_cuisine_unordered[cuisine]
    
    ingredient1,ingredient2 = pair.split("----")
    value+=cuisine_ingredients_paired_unordered[cuisine][ingredient1+"----"+ingredient2]
    value+=cuisine_ingredients_paired_unordered[cuisine][ingredient2+"----"+ingredient1]
    return 1.0*value/length
    
```

Ok, so I can find the probabilities for a particular cuisine and pair of ingredients. Now I need to sum. The list I have pays attention to order so I can't just sum there unless I take into account duplicates. Some orderings may be unique. Like, "tomato,egg" only happens once and so there is no "egg,tomato". So we can't just divide the whole sum by two. Ahh, but I can sum over every ingredient, ignoring when they are equal and then dividing by the number of pairs. "ingredients" is the variable that holds this information.

```python
measure=0
n=0
all_paired_ingredients_unordered=Counter()
for cuisine in cuisine_ingredients_paired_unordered:
    for pair in cuisine_ingredients_paired_unordered[cuisine]:
        all_paired_ingredients_unordered[pair]+=1

for cuisine1 in cuisine_ingredients_paired_unordered:
    print cuisine1
    for cuisine2 in cuisine_ingredients_paired_unordered:
        
        if cuisine1!=cuisine2:
            for pair,_ in all_paired_ingredients_unordered.most_common(30):
                x1=p_pair(cuisine1,pair)
                x2=p_pair(cuisine2,pair)
                measure+=abs(x1-x2)
                if x1!=0 or x2!=0: n+=1
            
    
measure = 1.0*measure/n
print measure
```
